##Horizontal bars ( barh() )

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y)  # the keyword argument height
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,color="red")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,color="Blue",height=0.1)
plt.show()
```

## Pie explode()

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([30,20,25,25])
y=["BMW","PORSCHE","POLO","GTR"]
z=[0.2,0,0,0]
plt.pie(x,labels=y,startangle=90,explode=z)
plt.legend()
plt.show()
```

```python

x=np.random.normal(170,10,4)
lb=["BMW","PORSCHE","POLO","GTR"]
ex=[.2,0,.2,0]
plt.pie(x,labels=lb,explode=ex)
plt.legend()
plt.show()
```

```python
c=np.array([180,90,45,45])
lb=["BMW","PORSCHE","POLO","GTR"]
ex=[.2,0,0,0]
color=['g','r',"b","k"]
plt.pie(c,labels=lb,explode=ex,colors=color,startangle=90)
plt.legend(title="CARS")
plt.show()
```

#Waffle

```python

import matplotlib.pyplot as plt
from pywaffle import Waffle
import numpy as np
```

```python
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=15,
    values=[50,10,10,30],
    figsize=(5,3),
    labels=["BMW","PORSCHE","POLO","GTR"],
    legend={'bbox_to_anchor':(0,0)}
)
plt.show()
```

```python
data={'democratic':48,'Republican':46,'Libertation':3}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values= data,
    legend={'bbox_to_anchor':(1,1)}
)
plt.show()
```

```python
data={'Democratic':48,'Republican':46,'Libertation':3}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values= data,
    colors=['#232066','#983D3D','#DCB732'],
    title={'label':'Vote Percentage In 2016 Election','loc':"left"},
    label=[f"{k}({v}%)"for k,v in data.items()],
    legend={'bbox_to_anchor':(1,1)} # to change the legend any position
)
fig.set_facecolor('#EEE')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
ypoint=np.array([34,24,56,12])
lb=["BMW","PORSCHE","POLO","GTR"]
ex=[.2,0,0,0]
plt.pie(ypoint,labels=lb,explode=ex,shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle
import numpy as np
data={'Democratic':48,'Republican':46,'Libertation':3}
fig=plt.figure(
    FigureClass=Waffle,
    rows=5,
    values=data,
    colors=['r','k','c'],
    legend={"bbox_to_anchor":(1,1)},
    icons="heart",
    font_size=10,
    icon_legend=True
)
plt.show()
```
